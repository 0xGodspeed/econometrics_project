```r
data <- read.csv("gwq_sdp_gini.csv")

# Run regression
model <- lm(gwq ~ sdp, data = data)

# Summary of the regression results
summary(model)
# Extract residuals from the model object
residuals <- residuals(model)

# Calculate sum of residuals
sum_of_residuals <- sum(residuals)

# View sum of residuals
print("Sum of residuals:")
print(sum_of_residuals)

# # Make predictions using the model
predicted <- predict(model)
# Plot residuals vs. sdp
# print mean squared error
print("MSE")
print(mean(residuals^2))
# print estimated coefficients
# print(model$coefficients)
```

```r
# Plot gwq vs actual sdp and predicted sdp
plot(data$sdp, data$gwq, col = "blue", pch = 20, xlab = "sdp", ylab = "gwq")
points(data$sdp, predicted, col = "red", pch = 20)
legend("topright", legend = c("Actual", "Predicted"), col = c("blue", "red"), pch = 20)
```

```r
# plot gwq vs sdp and show residuals
plot(data$sdp, data$gwq, col = "blue", pch = 20, xlab = "sdp", ylab = "gwq")
# make positive and negative residuals different colors
segments(data$sdp, data$gwq, data$sdp, predicted, col = ifelse(residuals > 0, "red", "green"))
# legend positive residuals in red and negative residuals in green
legend("topright", legend = c("Positive Residuals", "Negative Residuals"), col = c("red", "green"), lty = 1, lwd = 2)
```

```r
# plot residuals vs sdp 
plot(data$sdp, residuals, col = "blue", pch = 20, xlab = "sdp", ylab = "Residuals")
abline(h = 0, col = "red")
legend("topright", legend = c("Residuals", "Zero Residuals"), col = c("blue", "red"), pch = 20)
```

```r
# plot histogram of residuals, increase bins
hist(residuals, breaks = 500, col = "blue", xlab = "Residuals", main = "Histogram of Residuals")
```

```r
regions <- list(
  Northern = c('Chandigarh', 'Delhi', 'Haryana', 'Himachal Pradesh', 'Jammu and Kashmir', 'Ladakh', 'Punjab', 'Rajasthan'),
  `North-Eastern` = c('Arunachal Pradesh', 'Assam', 'Manipur', 'Meghalaya', 'Mizoram', 'Nagaland', 'Tripura'),
  Eastern = c('Andaman and Nicobar Islands', 'Bihar', 'Jharkhand', 'Odisha', 'Sikkim', 'West Bengal'), 
  Central = c('Chattisgarh', 'Madhya Pradesh', 'Uttar Pradesh', 'Uttarakhand'),
  Western = c('Dadra and Nagar Haveli', 'Daman and Diu', 'Goa', 'Gujarat', 'Maharashtra'),
  Southern = c('Andhra Pradesh', 'Karnataka', 'Kerala', 'Tamil Nadu', 'Telangana', 'Lakshadweep', 'Pondicherry')
)

# make new column in data called region
get_region <- function(state) {
  state <- tolower(state)
  for (region in names(regions)) {
    if (state %in% tolower(regions[[region]])) {
      return(region)
    }
  }
  return(NA)  # If state doesn't belong to any region
}

# Add 'region' column to data
data$region <- sapply(data$state, get_region)
head(data)
```

```r
# Run regression with region as a factor variable
model <- lm(gwq ~ sdp + I(sdp^2) + I(sdp^3) + gini + region, data = data)
summary(model)
print("Sum of residuals:")
print(sum_of_residuals)
# print mean squared error
print("Mean Squared Error")
print(mean(residuals^2))
```
